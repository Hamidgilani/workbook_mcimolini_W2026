```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("data/Property_Assessment_Data_(Current_Calendar_Year)_20251030.csv")

data.head()
```

```python
residential_data = data[data["Tax Class"] == "Residential"] # filters our data to where Tax Class == Residential
most_expensive = residential_data.sort_values(by="Assessed Value", ascending=False)[:6]

most_expensive.head(10)
```

```python
most_expensive["Property Address"] = most_expensive["House Number"].astype(str) + " " + most_expensive["Street Name"]
```

```python
# Set up storage for our data
property_names = most_expensive["Property Address"]
assessed_values = most_expensive["Assessed Value"]
```

```python
plt.figure(figsize=(10, 6))
plt.bar(property_names, assessed_values, color='skyblue')
plt.xlabel('Property Address')
plt.ylabel('Assessed Value')
plt.title('Top 6 Most Expensive Residential Properties')
plt.xticks(rotation=45, ha='right')  # Rotate x labels for better readability
plt.show()
```
