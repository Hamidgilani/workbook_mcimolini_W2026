```python
import pandas as pd
import matplotlib.pyplot as plt

# Set up my data
data = pd.read_csv("data/Property_Assessment_Data_(Current_Calendar_Year)_20251030.csv")

# Make sure it worked
data.head()
```

```python
residental_data = data[data["Tax Class"] == "Residential"] # gives us just our residential data

top_6_expensive = residental_data.sort_values(by="Assessed Value", ascending=False)[:6]

top_6_expensive.head(10) # will print as much as I have if there's less than requested
```

```python
top_6_expensive["Property Address"] = top_6_expensive["House Number"].astype(str) + " " + top_6_expensive["Street Name"]

# Placeholders for my plot
property_names = top_6_expensive["Property Address"]
assess_values = top_6_expensive["Assessed Value"]

# Plot it
plt.figure(figsize=(10, 6))
plt.bar(property_names, assess_values, color="green")
plt.xlabel("Property Address")
plt.ylabel("Value")
plt.title("Most Expensive Properties")
# xticks are the labels on your x-axis
plt.xticks(rotation=45, ha="right") # rotation rotates it, ha choses alignment
plt.show()
```

```python
# Plot it
plt.figure(figsize=(10, 6))
plt.plot(property_names, assess_values, marker="o", linestyle="-", color="teal")
plt.xlabel("Property Address")
plt.ylabel("Value")
plt.title("Most Expensive Properties")
# xticks are the labels on your x-axis
plt.xticks(rotation=45, ha="right") # rotation rotates it, ha choses alignment
plt.grid(True) # Enables grid-lines
plt.show()
```
